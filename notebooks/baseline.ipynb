```python
t_path = "D:\\Tesseract OCR\\tesseract.exe"
img_1 = cv2.imread("./data/train/DataTrain1.png")
brightandcontrast = cv2.convertScaleAbs(img_1, alpha=3.5, beta=50)
gray_image = cv2.cvtColor(brightandcontrast, cv2.COLOR_BGR2GRAY)
gray_image = cv2.bilateralFilter(gray_image, 11, 17, 17)
edged_image = cv2.Canny(gray_image, 30, 200)
```

```python
test_imgs = [f"./data/test/DataTest{i}.png" for i in range(1, 101)]

hasil = {}
detected = 0
for i, img in enumerate(test_imgs):
    img_1 = cv2.imread(img)
    gray_image = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
    gray_image = cv2.bilateralFilter(gray_image, 11, 17, 17)
    edged_image_1 = cv2.Canny(img_1, 30, 200)
    edged_image_2 = cv2.Canny(gray_image, 30, 200)
    
    text_ori = pytesseract.image_to_string(img_1)
    text_gray = pytesseract.image_to_string(gray_image)
    text_edge_1 = pytesseract.image_to_string(edged_image_1)
    text_edge_2 = pytesseract.image_to_string(edged_image_2)
    
    if text_ori != '' or text_gray != '' or text_edge_1 != '' or text_edge_2 != '':
        detected += 1
    
    hasil[i + 1] = [text_ori, text_gray, text_edge_1, text_edge_2]
    
    print( f"Image {i + 1} done" )
    print( f"Detected: {detected}" )
```
